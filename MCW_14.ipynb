Завдання 1. Побудувати множинну регресію для даних визначених нижче. У
відповіді написати рівняння регресії та R^2

image.png

Імпортуємо необхідні бібліотеки та таблицю.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = [
    [3, 67, 2008], [3, 72, 1960], [4, 77, 2022], [3, 65, 2013], [3, 58, 2020]
]

y = [869, 950, 564, 670, 443]

x, y = np.array(x), np.array(y)
```

Створюємо регресійну модель та навчаємо її.

```python
model = LinearRegression().fit(x, y)
```

Отримаємо результати.

```python
print('Відповідь:')
print()
r_sq = model.score(x, y)
print(f"Коефіцієнт детермінації R^2 = {r_sq}")
print()

intercept = model.intercept_
coef = model.coef_
print(f'Рівняння регресії:')
print()
print(f'Y = {round(intercept, 2)} {round(coef[0], 2)}*X1 + {round(coef[1], 2)}*X2 + {round(coef[2], 2)}*X3')
```

Завдання 2. В таблиці 1 визначені класи та атрибути. За допомогою методу
KNN визначити клас для запису з таблиці 2. Використовувати евклідову
метрику та обмежитися 3 сусідами.

image.png

image.png

Створимо Таблицю 1 та Таблицю 2.

```python
x = [
    [869, 3, 67.4, 2008], [950, 3, 72, 1960], [564, 4, 77.78, 2022], [496, 2, 41.57, 2020], [443, 3, 58.25, 2020]
]

y = [0, 3, 3, 4, 1]

x, y = np.array(x), np.array(y)

x_new = np.array([[670, 3, 65.1, 2013]])
```

Створюємо й навчаємо модель KNN.

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn.fit(x, y)
```

Визначаємо, до якого класу буде належати запис з Таблиці 2.

```python
print('Відповідь: ')
print()
prediction = knn.predict(x_new)
print("Клас для запису [670, 3, 65.1, 2013] з Таблиці 2 -", *prediction)
```
